Nestes exercícios, você deve decidir qual é o gráfico visto em aula que melhor visualiza uma base de dados. Após decidir, você deverá criar a visualização usando o conteúdo exposto durante a aula e adicionar um pequeno parágrafo sobre um insights que pode ser extraido do gráfico.

## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = sns.boxplot(x=data["price"])
```

**Insight do gráfico 1**: Diamantes raros vão custar mais de 12500 enquanto que a grande maioria poderá ser encontrada na faixa de 2500.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

diamonds = sns.scatterplot(data=data,x="carat", y = "price", hue ="clarity")
```

**Insight do gráfico 2**: Os produtos com a transparência "I1" são os que tem os diamantes mais pesados, porém em outros tipos de transparência com peso menor podemos encontrar diamantes na mesma faixa de preço.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = sns.histplot(data=data, x= "weight",kde=True)
```

**Insight do gráfico 3**: A maioria dos carros tem menos de 3 mil kilos e a medida que o peso aumenta vão se tornando mais "raros" em quantidade.
